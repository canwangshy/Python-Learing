**items()keys()values()**

items()keys()values()  key-value  key value dict_itemsdict_keys  dict_values Python  list() 

```python
cars = {'BMW': 8.5, 'BENS': 8.3, 'AUDI': 7.9}

# key-valuedict_items
ims = cars.items()
print(type(ims))

# dict_items
print(list(ims))

# 2key-value
print(list(ims)[1]) 

# keydict_keys
kys = cars.keys()
print(type(kys))

# dict_keys
print(list(kys))

# 2key
print(list(kys)[1])

# valuedict_values
vals = cars.values()

# dict_values
print(type(vals))

# 2value
print(list(vals)[1])
```
